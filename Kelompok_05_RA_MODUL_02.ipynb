```python
import pandas as pd
from functools import reduce

def load_data_from_file(file_path):
    try:
        data = pd.read_excel(file_path, header=None)
        data_list = data.squeeze().tolist()  # Mengubah DataFrame menjadi list
        return data_list
    except FileNotFoundError:
        print("File tidak ditemukan.")
        return None
    except Exception as e:
        print(f"Terjadi kesalahan saat membaca file: {e}")
        return None

def get_user_input():
    try:
        n = int(input("Masukkan jumlah data: "))
        data = []

        for i in range(n):
            value = float(input(f"Masukkan nilai ke-{i + 1}: "))
            data.append(value)
        return data

    except ValueError:
        print("Mohon masukkan angka yang valid.")
        return get_user_input()  # Ini mungkin membuat loop rekursif tanpa batas jika terus menerus salah input

def calculate_statistics(data):
    if not data:
        return None  # Menangani kasus ketika data kosong

    mean = lambda data: reduce(lambda x, y: x + y, data) / len(data)
    median = lambda data: sorted(data)[len(data) // 2] if len(data) % 2 != 0 else (sorted(data)[len(data) // 2 - 1] + sorted(data)[len(data) // 2]) / 2
    return {
        "max": max(data),
        "min": min(data),
        "mean": mean(data),
        "median": median(data)
    }

def main():
    while True:
        print("\n====Welcome to Program====")
        print("\nPilih data yang akan dimasukkan:")
        print("1. Input data")
        print("2. Upload file")
        print("3. Keluar")

        menu = input("\nMasukkan pilihan Anda (1/2/3): ")

        if menu == "3":
            print("\n====Sampai Jumpa====")
            break

        data = None
        if menu == "1":
            data = get_user_input()
        elif menu == "2":
            file_path = input("Masukkan file path: ")
            data = load_data_from_file(file_path)
        else:
            print("Pilihan tidak valid.")
            continue  # Melanjutkan loop while jika input tidak valid

        if data:
            # Menampilkan beberapa data awal
            print("\nData Awal:")
            print(data[:10])  # Menampilkan 10 data pertama jika panjang data melebihi 10
            stats = calculate_statistics(data)
            if stats:  # Menambahkan pemeriksaan jika stats adalah None
                print(stats)
            else:
                print("Data kosong atau tidak valid.")

if __name__ == "__main__":
    main()
4
```
